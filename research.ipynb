```python
df_revised = df_crovoc_labels.transpose().reset_index()
```

```python
df_revised = df_revised.rename(columns = {"index": "child"})
df_revised
```

```python
for i, p in enumerate(df_revised['parents']):
    #if type(p) is list:
    if len(p) > 0:
        df_revised['parents'][i] = p[0]
    else:
        df_revised = df_revised.drop(i) #makni one koje nemaju roditelje - samo ih je 3
        print(i)
```

```python
G = nx.from_pandas_edgelist(df_revised, source='parents', target='child', create_using=nx.DiGraph)
```

```python
roots = [node for node, degree in G.in_degree() if degree == 0]
roots
```

```python
leaves = [node for node, degree in G.out_degree() if degree == 0]
len(leaves)
```

```python
paths = []
for root in roots:
    for leaf in leaves:
        for path in nx.all_simple_paths(G, root, leaf):
            paths.append(path)


out = pd.DataFrame(paths).fillna('')
out.columns = reversed(out.add_prefix('level ').columns)
```
